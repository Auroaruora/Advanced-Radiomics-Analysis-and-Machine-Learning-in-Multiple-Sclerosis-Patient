```python
import pandas as pd
# This file generate GEMS_SNAPSHOT_Aurora.csv
# Key Changes

# 1. Rename the column 'Risk Score (1 low 2 high)' to 'Group', where:
# - 0 represents a healthy patient,
# - 1 represents a low-risk patient,
# - 2 represents a high-risk patient,
# - 3 represents an MS patient.
# Remove all the rows with empty entries (9 rows has been removed)

# Load the CSV file
data = pd.read_csv('GEMS_SNAPSHOT_Youzhi.csv')

# Clean the data
data = data.dropna(how='all')
data = data.dropna(axis=1, how='all')
data = data.drop('Merged Thalamic Nuclei Volumes/ICV corrected ', axis=1)

# Rename the column
data = data.rename(columns={'Risk Score (1 low 2 high)': 'Group'})

# Replace NaN values in the specified column with 0
data['Group'] = data['Group'].fillna(0)

# Ensure 'Subject' column is of string type
data['Subject'] = data['Subject'].astype(str)

# Updating the 'Group' column to 3 for subjects not starting with 'P'
data['Group'] = data.apply(lambda row: 3 if row['Subject'].startswith('9') else row['Group'], axis=1)

# Dropping rows with any missing values
data = data.dropna()

# Converting all columns except 'Subject' to numerical values and removing rows with invalid entries
cols = data.columns.drop('Subject')
data[cols] = data[cols].apply(pd.to_numeric, errors='coerce')
data = data.dropna()
```
